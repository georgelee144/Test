```python
import pandas as pd
import json
from collections import defaultdict

file = 'data/world_bank_projects.json'

with open(file) as json_file:
    json_data = json.load(json_file)

df = pd.DataFrame(json_data)

df_count=defaultdict(int)

for countries in df['countryname']:
    if countries == df_count[countries]:
        df_count[countries] += 1
    else:
        df_count[countries] = 1

t = 0
for keys in df_count.keys():
    t+=1
print(t)
```
